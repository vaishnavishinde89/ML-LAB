```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (adjust path if needed)
df = pd.read_csv("iris (1).csv")

# Clean column names
df.columns = [c.strip().lower().replace(' ', '_') for c in df.columns]

print(df.head())
print(df['species'].value_counts())
```

```python
species_counts = df['species'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Species Distribution (Pie Chart)")
plt.show()
```

```python
species_counts.plot(kind='bar', color=['skyblue','lightgreen','salmon'])
plt.title("Species Count (Bar Chart)")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

```python
df.hist(figsize=(10,8), bins=15, edgecolor="black")
plt.suptitle("Histograms of all features")
plt.show()
```

```python
# Shape of dataset
print("Shape:", df.shape)

# Column names and data types
print("\nInfo:")
print(df.info())

# Summary statistics
print("\nDescribe:")
print(df.describe())

# Check missing values
print("\nMissing values:")
print(df.isnull().sum())
```

```python
df['sepal_length'] = df['sepal_length'].fillna(df['sepal_length'].mean())
```

```python
# Count duplicate rows
print("Duplicate rows:", df.duplicated().sum())

# Drop duplicate rows
df = df.drop_duplicates()
```
